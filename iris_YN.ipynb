```python
# כאן למעלה ניתן לראות עוד גרף סקטרפלוט שהפעם נעשה באצמעות SEABORN
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='species', y='sepal_width', data=iris, errorbar=None)
plt.title('Mean Sepal Width for Each Species')
plt.xlabel('Species')
plt.ylabel('Mean Sepal Width')
plt.show()
```

```python
# גרף העמודות הנ"ל מציג את ממוצע הרוחב של עלי הגביע. אפשר לראות שסטורה הוא הרחב ביותר, אחריו וירג'יניקה ואחרון ורסיקולור.
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='sepal_width', data=iris)
plt.title('Distribution of Sepal Width by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Width')
plt.show()
```

```python
# בגרף הקופסא הנ"ל אנחנו רואים את ההתפלגות של המשתנה רוחב עלי הגביע בכל זן אירוס 
#  אפשר לראות שסטורה הוא הרחב ביותר, אחריו וירג'יניקה ואחרון ורסיקולור.
# באופן כללי, ניתן לראות שיש הבדלים מובהקים בין המינים מבחינת רוחב עלי הכותרת, כאשר לסטורה יש רוחב עלי כותרת גדול יותר ופיזור נתונים רחב יותר בהשוואה לשני המינים האחרים.
#  סטורה בעל חציון גבוה יותר (בערך 3.4) בהשוואה לשני המינים האחרים להם חציון בערך 2.8 -3 
```

```python
# Create a scatter plot with Seaborn
sns.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species')
plt.title('Scatter Plot of Sepal Length vs. Petal Length by Species')
# Show the plot
plt.show()
```

```python
#הגרף מציג את הקשר בין אורך הגביע לבין אורך עלי הכותרת בשלושה מיני אירוס.
#ניתן לראות הבדל ברור בין המינים: המין סטורה מאופיין באורך גביע ועלי כותרת קטנים יותר, בעוד שהמינים ורסיקולור וורג'יניקה נבדלים זה מזה באורכים הגדולים יותר של עלי הכותרת, כאשר וירג'יניקה מאופיין באורכים הגדולים ביותר.
#באופן כללי אפשר לראות שבמינים הללו יש קשר חיובי בין המשתנים ובסוטרה אורך עלי הגביע אינו בהכרח עולה בקנה מידה עם אורך עלי הכותרת.
```

```python
g = sns.relplot(data=iris, x="sepal_width", y="petal_width", col="species")
g.set_axis_labels("Sepal Width", "Petal Width")
plt.tight_layout()
plt.show()
```
